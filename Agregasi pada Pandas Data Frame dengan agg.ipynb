```python
#PERSIAPAN DATA FRAME
df = pd.read_csv('/content/Iris.csv') #Sebagai Data Frame
df
```

```python
#Mengenal groupby() dan fungsi agregasi
df.groupby('Species')['PetalLengthCm'].count().to_frame() #Untuk menghitung jumlah baris yang ditampilkan sebagai frame
```

```python
df.groupby('Species')['PetalLengthCm'].mean().to_frame() #Untuk menghitung rata-rata yang ditampilkan sebagai frame
```

```python
df.groupby('Species')['PetalLengthCm'].median().to_frame() #Untuk menghitung nilai tengah yang ditampilkan sebagai frame
```

```python
#Agregasi dengan agg()

df.groupby('Species')['PetalLengthCm'].agg(['count', 'mean', 'median']) #Untuk menggunakan fungsi agregasi secara langsung dan tidak terpisah
```

```python
df.groupby('Species')['PetalLengthCm'].describe() #Untuk menampilkan deskriptif statistik
```
